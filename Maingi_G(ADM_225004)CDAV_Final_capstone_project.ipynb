```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Setting up visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12
custom_palette = ["#2E86AB", "#A23B72", "#F18F01", "#C73E1D", "#4CAF50", "#FFC107"]

# Load the dataset
df = pd.read_csv('health_ken.csv')

# Display the first & last 10 rows
print("Dataset Head:")
print(df.head(10))
print(df.tail(10))

# Check dataset structure
print("\nDataset Info:")
df.info()

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Step 1: Data Cleaning and Preparation
# Remove metadata rows (first two rows)
df_clean = df.iloc[2:].reset_index(drop=True)

# Rename columns for clarity
df_clean.columns = [ 'Country', 'Country_Code', 'Year', 'Indicator_Name', 'Indicator_Code', 'Value']

# Convert data types
df_clean['Year'] = pd.to_numeric(df_clean['Year'])
df_clean['Value'] = pd.to_numeric(df_clean['Value'], errors='coerce')

# Filter for Kenya only
df_kenya = df_clean[df_clean['Country_Code'] == 'KEN'].copy()

# Handle missing values - forward fill within each indicator group
df_kenya['Value'] = df_kenya.groupby('Indicator_Code')['Value'].transform(
    lambda x: x.fillna(method='ffill').fillna(method='bfill')
)

# Create new columns for analysis
# 1. Decade column
df_kenya['Decade'] = (df_kenya['Year'] // 10) * 10

# 2. Health category mapping
def map_health_category(code):
    if code.startswith('SG.VAW'):
        return 'Gender-Based Violence'
    elif code.startswith('SH.ALC'):
        return 'Alcohol Consumption'
    elif code.startswith('SH.ANM'):
        return 'Anemia Prevalence'
    elif code.startswith('SH.DTH'):
        return 'Mortality Statistics'
    elif code.startswith('SH.DYN'):
        return 'HIV/AIDS'
    else:
        return 'Other Health Indicators'

df_kenya['Health_Category'] = df_kenya['Indicator_Code'].apply(map_health_category)

# 3. Demographic group mapping
def map_demographic(indicator):
    indicator_lower = indicator.lower()
    if 'children' in indicator_lower:
        return 'Children'
    elif 'women' in indicator_lower:
        return 'Women'
    elif 'infant' in indicator_lower or 'neonatal' in indicator_lower:
        return 'Infants'
    elif 'male' in indicator_lower:
        return 'Men'
    elif 'female' in indicator_lower:
        return 'Female'
    else:
        return 'General Population'

df_kenya['Demographic_Group'] = df_kenya['Indicator_Name'].apply(map_demographic)

# 4. Time period mapping
def map_time_period(year):
    if year < 1990:
        return 'Pre-1990'
    elif 1990 <= year < 2000:
        return '1990s'
    elif 2000 <= year < 2010:
        return '2000s'
    elif 2010 <= year < 2020:
        return '2010s'
    else:
        return '2020s'

df_kenya['Time_Period'] = df_kenya['Year'].apply(map_time_period)

# Save cleaned data
df_kenya.to_csv('kenya_health_cleaned.csv', index=False)

# Step 2: Exploratory Data Analysis
# Pivot data for analysis
df_pivot = df_kenya.pivot_table(
    index=['Year', 'Decade', 'Health_Category', 'Time_Period'],
    columns='Indicator_Code',
    values='Value'
).reset_index()

# Rename key columns
key_columns = {
    'SG.VAW.1549.ZS': 'Violence_Against_Women',
    'SH.ALC.PCAP.FE.LI': 'Alcohol_Consumption_Female',
    'SH.ALC.PCAP.LI': 'Alcohol_Consumption_Total',
    'SH.ALC.PCAP.MA.LI': 'Alcohol_Consumption_Male',
    'SH.ANM.ALLW.ZS': 'Anemia_Women',
    'SH.ANM.CHLD.ZS': 'Anemia_Children',
    'SH.ANM.NPRG.ZS': 'Anemia_NonPregnant_Women',
    'SH.DTH.IMRT': 'Infant_Deaths',
    'SH.DTH.MORT': 'UnderFive_Deaths',
    'SH.DYN.AIDS.ZS': 'HIV_Prevalence',
    'SH.DYN.MORT': 'UnderFive_Mortality_Rate',
    'SH.CON.1524.FE.ZS': 'Condom_Use_Female',
    'SH.CON.1524.MA.ZS': 'Condom_Use_Male',
    'SH.DYN.AIDS.FE.ZS': 'HIV_Women_Share'
}
df_pivot = df_pivot.rename(columns=key_columns)

# Summary statistics
print("\nSummary Statistics:")
print(df_pivot.describe())

# Health category distribution
plt.figure(figsize=(10, 6))
health_counts = df_kenya['Health_Category'].value_counts()
sns.barplot(x=health_counts.values, y=health_counts.index, palette=custom_palette)
plt.title('Distribution of Health Indicators by Category')
plt.xlabel('Number of Records')
plt.tight_layout()
plt.savefig('health_category_distribution.png')
plt.show()

# Time period distribution
plt.figure(figsize=(10, 6))
time_counts = df_kenya['Time_Period'].value_counts()
sns.barplot(x=time_counts.index, y=time_counts.values, palette=custom_palette)
plt.title('Distribution of Data by Time Period')
plt.xlabel('Time Period')
plt.ylabel('Number of Records')
plt.tight_layout()
plt.savefig('time_period_distribution.png')
plt.show()

# Year distribution
plt.figure(figsize=(12, 6))
sns.histplot(df_kenya['Year'], bins=30, kde=True, color="#2E86AB")
plt.title('Distribution of Data by Year')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('year_distribution.png')
plt.show()

# Step 3: Time Series Analysis
# Mortality trends
mortality_df = df_pivot[['Year', 'UnderFive_Mortality_Rate']].dropna()
plt.figure(figsize=(12, 6))
plt.plot(mortality_df['Year'], mortality_df['UnderFive_Mortality_Rate'], marker='o', color="#C73E1D")
plt.title('Under-5 Mortality Rate in Kenya (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Rate per 1,000 live births')
plt.grid(True)
plt.tight_layout()
plt.savefig('mortality_trend.png')
plt.show()

# HIV prevalence trend
hiv_df = df_pivot[['Year', 'HIV_Prevalence']].dropna()
plt.figure(figsize=(12, 6))
plt.fill_between(hiv_df['Year'], hiv_df['HIV_Prevalence'], color="#A23B72", alpha=0.3)
plt.plot(hiv_df['Year'], hiv_df['HIV_Prevalence'], marker='o', color="#A23B72")
plt.title('HIV Prevalence in Kenya (1990-2023)')
plt.xlabel('Year')
plt.ylabel('Prevalence (%)')
plt.grid(True)
plt.tight_layout()
plt.savefig('hiv_trend.png')
plt.show()

# Anemia comparison across all groups
anemia_df = df_pivot[['Year', 'Anemia_Women', 'Anemia_Children', 'Anemia_NonPregnant_Women']].dropna()
plt.figure(figsize=(12, 6))
plt.plot(anemia_df['Year'], anemia_df['Anemia_Women'], marker='o', label='Women', color="#A23B72")
plt.plot(anemia_df['Year'], anemia_df['Anemia_Children'], marker='s', label='Children', color="#F18F01")
plt.plot(anemia_df['Year'], anemia_df['Anemia_NonPregnant_Women'], marker='^', label='Non-Pregnant Women', color="#4CAF50")
plt.title('Anemia Prevalence by Demographic Group in Kenya')
plt.xlabel('Year')
plt.ylabel('Prevalence (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('anemia_comparison_all_groups.png')
plt.show()

# Alcohol consumption trends by gender
alcohol_df = df_pivot[['Year', 'Alcohol_Consumption_Male', 'Alcohol_Consumption_Female', 'Alcohol_Consumption_Total']].dropna()
plt.figure(figsize=(12, 6))
plt.plot(alcohol_df['Year'], alcohol_df['Alcohol_Consumption_Male'], marker='o', label='Male', color="#2E86AB")
plt.plot(alcohol_df['Year'], alcohol_df['Alcohol_Consumption_Female'], marker='s', label='Female', color="#A23B72")
plt.plot(alcohol_df['Year'], alcohol_df['Alcohol_Consumption_Total'], marker='^', label='Total', color="#F18F01")
plt.title('Alcohol Consumption by Gender in Kenya')
plt.xlabel('Year')
plt.ylabel('Liters per capita')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('alcohol_consumption_by_gender.png')
plt.show()

# Condom use by gender
condom_df = df_pivot[['Year', 'Condom_Use_Male', 'Condom_Use_Female']].dropna()
plt.figure(figsize=(12, 6))
plt.plot(condom_df['Year'], condom_df['Condom_Use_Male'], marker='o', label='Male', color="#2E86AB")
plt.plot(condom_df['Year'], condom_df['Condom_Use_Female'], marker='s', label='Female', color="#A23B72")
plt.title('Condom Use by Gender in Kenya (15-24 years)')
plt.xlabel('Year')
plt.ylabel('Usage (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('condom_use_by_gender.png')
plt.show()

# HIV gender share
hiv_gender_df = df_pivot[['Year', 'HIV_Women_Share']].dropna()
plt.figure(figsize=(12, 6))
plt.plot(hiv_gender_df['Year'], hiv_gender_df['HIV_Women_Share'], marker='o', color="#A23B72")
plt.title('Women\'s Share of HIV Population in Kenya (15+ years)')
plt.xlabel('Year')
plt.ylabel('Share (%)')
plt.grid(True)
plt.tight_layout()
plt.savefig('hiv_gender_share.png')
plt.show()

# Step 4: Comparative Analysis
# Health indicators by time period
plt.figure(figsize=(12, 6))
time_period_df = df_pivot.groupby('Time_Period')[['HIV_Prevalence', 'Anemia_Women', 'UnderFive_Mortality_Rate']].mean()
time_period_df.plot(kind='bar', figsize=(12, 6), color=custom_palette[:3])
plt.title('Health Indicators by Time Period')
plt.xlabel('Time Period')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('indicators_by_period.png')
plt.show()

# Gender-based violence analysis
violence_cols = [col for col in df_pivot.columns if 'Violence' in col and col != 'Violence_Against_Women']
if violence_cols:
    violence_df = df_pivot[['Year'] + violence_cols].dropna()
    plt.figure(figsize=(12, 6))
    for col in violence_cols:
        plt.plot(violence_df['Year'], violence_df[col], marker='o', label=col.replace('Violence_', '').replace('_', ' '))
    plt.title('Trends in Violence Justification Among Women in Kenya')
    plt.xlabel('Year')
    plt.ylabel('Percentage of Women Who Justify Violence')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('violence_trends.png')
    plt.show()

# Step 5: Correlation Analysis
# Select key indicators for correlation analysis
key_indicators = ['UnderFive_Mortality_Rate', 'HIV_Prevalence', 'Alcohol_Consumption_Male',
                 'Alcohol_Consumption_Female', 'Anemia_Women', 'Anemia_Children', 'Anemia_NonPregnant_Women']
corr_df = df_pivot[key_indicators].dropna()

# Calculate correlation matrix
corr_matrix = corr_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,
            linewidths=0.5, linecolor='white', fmt='.2f')
plt.title('Correlation Matrix of Key Health Indicators')
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.show()

# Step 6: Hypothesis Testing
print("\n=== HYPOTHESIS TESTING ===\n")

# Hypothesis 1: Under-five mortality has significantly decreased since 2000
print("HYPOTHESIS 1: Under-five mortality has significantly decreased since 2000")
print("Null Hypothesis (H0): There is no significant difference in under-five mortality between 2000-2005 and 2018-2023")
print("Alternative Hypothesis (H1): Under-five mortality has significantly decreased since 2000")

early_2000s = df_pivot[(df_pivot['Year'] >= 2000) & (df_pivot['Year'] <= 2005)]['UnderFive_Mortality_Rate'].dropna()
recent = df_pivot[(df_pivot['Year'] >= 2018) & (df_pivot['Year'] <= 2023)]['UnderFive_Mortality_Rate'].dropna()

if len(early_2000s) > 0 and len(recent) > 0:
    t_stat, p_val = stats.ttest_ind(early_2000s, recent)
    print(f"\nT-statistic: {t_stat:.3f}")
    print(f"P-value: {p_val:.5f}")

    if p_val < 0.05:
        print("Decision: Reject the null hypothesis (H0)")
        print("Conclusion: There is sufficient evidence to support that under-five mortality has significantly decreased since 2000")
    else:
        print("Decision: Fail to reject the null hypothesis (H0)")
        print("Conclusion: There is insufficient evidence to support that under-five mortality has significantly decreased since 2000")
else:
    print("Insufficient data for this hypothesis test")

# Hypothesis 2: HIV prevalence has significantly decreased since its peak
print("\n\nHYPOTHESIS 2: HIV prevalence has significantly decreased since its peak")
print("Null Hypothesis (H0): There is no significant difference in HIV prevalence between peak years (1995-2000) and recent years (2018-2023)")
print("Alternative Hypothesis (H1): HIV prevalence has significantly decreased since its peak")

hiv_peak = df_pivot[(df_pivot['Year'] >= 1995) & (df_pivot['Year'] <= 2000)]['HIV_Prevalence'].dropna()
hiv_recent = df_pivot[(df_pivot['Year'] >= 2018) & (df_pivot['Year'] <= 2023)]['HIV_Prevalence'].dropna()

if len(hiv_peak) > 0 and len(hiv_recent) > 0:
    t_stat, p_val = stats.ttest_ind(hiv_peak, hiv_recent)
    print(f"\nT-statistic: {t_stat:.3f}")
    print(f"P-value: {p_val:.5f}")

    if p_val < 0.05:
        print("Decision: Reject the null hypothesis (H0)")
        print("Conclusion: There is sufficient evidence to support that HIV prevalence has significantly decreased since its peak")
    else:
        print("Decision: Fail to reject the null hypothesis (H0)")
        print("Conclusion: There is insufficient evidence to support that HIV prevalence has significantly decreased since its peak")
else:
    print("Insufficient data for this hypothesis test")

# Hypothesis 3: Anemia prevalence differs significantly between demographic groups
print("\n\nHYPOTHESIS 3: Anemia prevalence differs significantly between demographic groups")
print("Null Hypothesis (H0): There is no significant difference in anemia prevalence between demographic groups")
print("Alternative Hypothesis (H1): Anemia prevalence differs significantly between demographic groups")

women_anemia = df_pivot['Anemia_Women'].dropna()
children_anemia = df_pivot['Anemia_Children'].dropna()
non_pregnant_anemia = df_pivot['Anemia_NonPregnant_Women'].dropna()

# Compare women vs children
if len(women_anemia) > 0 and len(children_anemia) > 0:
    t_stat, p_val = stats.ttest_ind(women_anemia, children_anemia)
    print(f"\nWomen vs Children - T-statistic: {t_stat:.3f}, P-value: {p_val:.5f}")
    if p_val < 0.05:
        print("Decision: Reject the null hypothesis (H0)")
        print("Conclusion: Significant difference between women and children")
    else:
        print("Decision: Fail to reject the null hypothesis (H0)")
        print("Conclusion: No significant difference between women and children")

# Compare pregnant vs non-pregnant women
if len(women_anemia) > 0 and len(non_pregnant_anemia) > 0:
    t_stat, p_val = stats.ttest_ind(women_anemia, non_pregnant_anemia)
    print(f"\nPregnant vs Non-Pregnant Women - T-statistic: {t_stat:.3f}, P-value: {p_val:.5f}")
    if p_val < 0.05:
        print("Decision: Reject the null hypothesis (H0)")
        print("Conclusion: Significant difference between pregnant and non-pregnant women")
    else:
        print("Decision: Fail to reject the null hypothesis (H0)")
        print("Conclusion: No significant difference between pregnant and non-pregnant women")
else:
    print("Insufficient data for this comparison")

# Hypothesis 4: Alcohol consumption differs significantly between genders
print("\n\nHYPOTHESIS 4: Alcohol consumption differs significantly between genders")
print("Null Hypothesis (H0): There is no significant difference in alcohol consumption between males and females")
print("Alternative Hypothesis (H1): Alcohol consumption differs significantly between males and females")

male_alcohol = df_pivot['Alcohol_Consumption_Male'].dropna()
female_alcohol = df_pivot['Alcohol_Consumption_Female'].dropna()

if len(male_alcohol) > 0 and len(female_alcohol) > 0:
    t_stat, p_val = stats.ttest_ind(male_alcohol, female_alcohol)
    print(f"\nT-statistic: {t_stat:.3f}")
    print(f"P-value: {p_val:.5f}")

    if p_val < 0.05:
        print("Decision: Reject the null hypothesis (H0)")
        print("Conclusion: There is sufficient evidence to support that alcohol consumption differs significantly between genders")
    else:
        print("Decision: Fail to reject the null hypothesis (H0)")
        print("Conclusion: There is insufficient evidence to support that alcohol consumption differs significantly between genders")
else:
    print("Insufficient data for this hypothesis test")

# Hypothesis 5: ANOVA - Health indicators differ across decades
print("\n\nHYPOTHESIS 5: Health indicators differ across decades")
print("Null Hypothesis (H0): There is no significant difference in health indicators across decades")
print("Alternative Hypothesis (H1): Health indicators differ significantly across decades")

decade_data = df_pivot.groupby('Decade')['UnderFive_Mortality_Rate'].apply(list).dropna()

# Check if we have at least 2 decades with data
if len(decade_data) >= 2:
    # Check if each decade has at least 2 data points
    valid_decades = [decade for decade in decade_data if len(decade) >= 2]
    if len(valid_decades) >= 2:
        f_stat, p_val = stats.f_oneway(*valid_decades)
        print(f"\nF-statistic: {f_stat:.3f}")
        print(f"P-value: {p_val:.5f}")

        if p_val < 0.05:
            print("Decision: Reject the null hypothesis (H0)")
            print("Conclusion: There is sufficient evidence to support that health indicators differ significantly across decades")
        else:
            print("Decision: Fail to reject the null hypothesis (H0)")
            print("Conclusion: There is insufficient evidence to support that health indicators differ significantly across decades")
    else:
        print("Insufficient data points in each decade for ANOVA")
else:
    print("Insufficient decades for ANOVA (need at least 2)")

# Visualization of hypothesis test results
plt.figure(figsize=(14, 10))

# 1. Mortality comparison
plt.subplot(2, 2, 1)
if len(early_2000s) > 0 and len(recent) > 0:
    plt.boxplot([early_2000s, recent], labels=['2000-2005', '2018-2023'])
    plt.title('Under-5 Mortality Rate Comparison')
    plt.ylabel('Rate per 1,000 live births')
else:
    plt.text(0.5, 0.5, 'Insufficient Data', ha='center', va='center')
    plt.title('Under-5 Mortality Rate Comparison')

# 2. HIV prevalence comparison
plt.subplot(2, 2, 2)
if len(hiv_peak) > 0 and len(hiv_recent) > 0:
    plt.boxplot([hiv_peak, hiv_recent], labels=['Peak (1995-2000)', 'Recent (2018-2023)'])
    plt.title('HIV Prevalence Comparison')
    plt.ylabel('Prevalence (%)')
else:
    plt.text(0.5, 0.5, 'Insufficient Data', ha='center', va='center')
    plt.title('HIV Prevalence Comparison')

# 3. Anemia comparison
plt.subplot(2, 2, 3)
if len(women_anemia) > 0 and len(children_anemia) > 0 and len(non_pregnant_anemia) > 0:
    plt.boxplot([women_anemia, children_anemia, non_pregnant_anemia],
                labels=['Women', 'Children', 'Non-Pregnant Women'])
    plt.title('Anemia Prevalence Comparison')
    plt.ylabel('Prevalence (%)')
else:
    plt.text(0.5, 0.5, 'Insufficient Data', ha='center', va='center')
    plt.title('Anemia Prevalence Comparison')

# 4. Decade comparison
plt.subplot(2, 2, 4)
decade_means = df_pivot.groupby('Decade')['UnderFive_Mortality_Rate'].mean()
if len(decade_means) > 0:
    plt.bar(decade_means.index, decade_means.values, color=custom_palette[:len(decade_means)])
    plt.title('Average Under-5 Mortality by Decade')
    plt.xlabel('Decade')
    plt.ylabel('Average Rate')
else:
    plt.text(0.5, 0.5, 'Insufficient Data', ha='center', va='center')
    plt.title('Average Under-5 Mortality by Decade')

plt.tight_layout()
plt.savefig('hypothesis_test_results.png')
plt.show()

# Step 7: Comprehensive Dashboard Visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Kenya Health Progress Dashboard (2000-2023)', fontsize=16)

# 1. Mortality Trends
axes[0, 0].plot(mortality_df['Year'], mortality_df['UnderFive_Mortality_Rate'], marker='o', color="#C73E1D")
axes[0, 0].set_title('Under-5 Mortality Rate')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Rate per 1,000 live births')
axes[0, 0].grid(True)

# 2. HIV Prevalence
axes[0, 1].fill_between(hiv_df['Year'], hiv_df['HIV_Prevalence'], color="#A23B72", alpha=0.3)
axes[0, 1].plot(hiv_df['Year'], hiv_df['HIV_Prevalence'], marker='o', color="#A23B72")
axes[0, 1].set_title('HIV Prevalence')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Prevalence (%)')
axes[0, 1].grid(True)

# 3. Anemia Comparison
anemia_df = df_pivot[['Year', 'Anemia_Women', 'Anemia_Children', 'Anemia_NonPregnant_Women']].dropna()
axes[0, 2].plot(anemia_df['Year'], anemia_df['Anemia_Women'], marker='o', label='Women', color="#A23B72")
axes[0, 2].plot(anemia_df['Year'], anemia_df['Anemia_Children'], marker='s', label='Children', color="#F18F01")
axes[0, 2].plot(anemia_df['Year'], anemia_df['Anemia_NonPregnant_Women'], marker='^', label='Non-Pregnant Women', color="#4CAF50")
axes[0, 2].set_title('Anemia Prevalence by Group')
axes[0, 2].set_xlabel('Year')
axes[0, 2].set_ylabel('Prevalence (%)')
axes[0, 2].legend()
axes[0, 2].grid(True)

# 4. Alcohol Consumption by Gender
alcohol_df = df_pivot[['Year', 'Alcohol_Consumption_Male', 'Alcohol_Consumption_Female']].dropna()
axes[1, 0].plot(alcohol_df['Year'], alcohol_df['Alcohol_Consumption_Male'], marker='o', label='Male', color="#2E86AB")
axes[1, 0].plot(alcohol_df['Year'], alcohol_df['Alcohol_Consumption_Female'], marker='s', label='Female', color="#A23B72")
axes[1, 0].set_title('Alcohol Consumption by Gender')
axes[1, 0].set_xlabel('Year')
axes[1, 0].set_ylabel('Liters per capita')
axes[1, 0].legend()
axes[1, 0].grid(True)

# 5. Health Category Distribution - ENLARGED PIE CHART
health_counts = df_kenya['Health_Category'].value_counts()
axes[1, 1].pie(health_counts.values, labels=health_counts.index, autopct='%1.1f%%',
                colors=custom_palette, textprops={'fontsize': 12},
                explode=[0.05] * len(health_counts), radius=1.2)
axes[1, 1].set_title('Health Indicators Distribution', fontsize=14)

# 6. Correlation Heatmap
key_indicators = ['UnderFive_Mortality_Rate', 'HIV_Prevalence', 'Alcohol_Consumption_Male',
                 'Anemia_Women', 'Anemia_Children', 'Anemia_NonPregnant_Women']
corr_df = df_pivot[key_indicators].dropna()
corr_matrix = corr_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,
            linewidths=0.5, linecolor='white', fmt='.2f', ax=axes[1, 2])
axes[1, 2].set_title('Correlation Matrix')

plt.tight_layout()
plt.savefig('health_dashboard.png', dpi=300)
plt.show()

# Step 8: Key Insights Summary
print("\n=== KEY HEALTH INSIGHTS FOR KENYA ===\n")

# Calculate key metrics if data is available
if 2000 in df_pivot['Year'].values and 2023 in df_pivot['Year'].values:
    mortality_2000 = df_pivot[df_pivot['Year'] == 2000]['UnderFive_Mortality_Rate'].values[0]
    mortality_2023 = df_pivot[df_pivot['Year'] == 2023]['UnderFive_Mortality_Rate'].values[0]
    mortality_reduction = ((mortality_2000 - mortality_2023) / mortality_2000) * 100
    print(f"1. Under-5 Mortality Reduction: {mortality_reduction:.1f}% from 2000 to 2023")
else:
    print("1. Under-5 Mortality data not available for both 2000 and 2023")

if 1996 in df_pivot['Year'].values and 2023 in df_pivot['Year'].values:
    hiv_peak_year = df_pivot[df_pivot['Year'] == 1996]['HIV_Prevalence'].values[0]
    hiv_2023 = df_pivot[df_pivot['Year'] == 2023]['HIV_Prevalence'].values[0]
    hiv_reduction = ((hiv_peak_year - hiv_2023) / hiv_peak_year) * 100
    print(f"2. HIV Prevalence Reduction: {hiv_reduction:.1f}% from peak (1996) to 2023")
else:
    print("2. HIV Prevalence data not available for both peak year and 2023")

# Anemia insights across all groups
if not df_pivot['Anemia_Women'].isna().all() and not df_pivot['Anemia_Children'].isna().all() and not df_pivot['Anemia_NonPregnant_Women'].isna().all():
    women_anemia = df_pivot['Anemia_Women'].mean()
    children_anemia = df_pivot['Anemia_Children'].mean()
    non_pregnant_anemia = df_pivot['Anemia_NonPregnant_Women'].mean()

    print(f"3. Anemia Prevalence by Group:")
    print(f"   - Women: {women_anemia:.1f}%")
    print(f"   - Children: {children_anemia:.1f}%")
    print(f"   - Non-Pregnant Women: {non_pregnant_anemia:.1f}%")
    print(f"   - Gap between pregnant and non-pregnant women: {abs(women_anemia - non_pregnant_anemia):.1f}%")
else:
    print("3. Anemia data not available for all groups")

# Alcohol consumption by gender
if not df_pivot['Alcohol_Consumption_Male'].isna().all() and not df_pivot['Alcohol_Consumption_Female'].isna().all():
    male_alcohol = df_pivot['Alcohol_Consumption_Male'].mean()
    female_alcohol = df_pivot['Alcohol_Consumption_Female'].mean()
    gender_gap = male_alcohol - female_alcohol
    print(f"4. Alcohol Consumption by Gender:")
    print(f"   - Male: {male_alcohol:.2f} liters per capita")
    print(f"   - Female: {female_alcohol:.2f} liters per capita")
    print(f"   - Gender Gap: {gender_gap:.2f} liters per capita")
else:
    print("4. Alcohol Consumption data not available for both genders")

# Women's share of HIV population
if not df_pivot['HIV_Women_Share'].isna().all():
    hiv_women_share = df_pivot['HIV_Women_Share'].mean()
    print(f"5. Women's Share of HIV Population: {hiv_women_share:.1f}%")
    print(f"   This indicates women bear a disproportionate burden of HIV in Kenya")
else:
    print("5. HIV gender share data not available")

if not df_pivot['UnderFive_Mortality_Rate'].isna().all():
    best_decade = df_pivot.groupby('Decade')['UnderFive_Mortality_Rate'].mean().idxmin()
    print(f"6. Best Health Decade: {best_decade}s (lowest average under-5 mortality)")
else:
    print("6. Under-5 Mortality data not available for decade analysis")

print("\n=== RECOMMENDATIONS ===\n")
print("1. Continue strengthening maternal and child health programs")
print("2. Maintain HIV prevention efforts with focus on women who bear disproportionate burden")
print("3. Implement targeted nutrition interventions for all women, with special attention to pregnant women")
print("4. Monitor alcohol consumption trends with gender-specific interventions")
print("5. Focus on integrated health interventions addressing all demographic groups")
print("6. Address gender disparities in health outcomes across all population segments")
print("7. Learn from successful decades to improve future health outcomes")

print("\n=== ANALYSIS COMPLETE ===")
```
