```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# Load and preprocess the data
def load_and_preprocess():
    # Convert the provided CSV data to a DataFrame
    data = """
0 | Date | Time | Booking ID | Booking Status | Customer ID | Vehicle Type | Pickup Location | Drop Location | Avg VTAT | Avg CTAT | Cancelled Rides by Customer | Reason for cancelling by Customer | Cancelled Rides by Driver | Driver Cancellation Reason | Incomplete Rides | Incomplete Rides Reason | Booking Value | Ride Distance | Driver Ratings | Customer Rating | Payment Method
1 | 2024-03-23 | 12:29:38 | "CNR5884300" | No Driver Found | "CID1982111" | eBike | Palam Vihar | Jhilmil | null | null | null | null | null | null | null | null | null | null | null | null | null
2 | 2024-11-29 | 18:01:39 | "CNR1326809" | Incomplete | "CID4604802" | Go Sedan | Shastri Nagar | Gurgaon Sector 56 | 4.9 | 14.0 | null | null | null | null | 1 | Vehicle Breakdown | 237 | 5.73 | null | null | UPI
3 | 2024-08-23 | 08:56:10 | "CNR8494506" | Completed | "CID9202816" | Auto | Khandsa | Malviya Nagar | 13.4 | 25.8 | null | null | null | null | null | null | 627 | 13.58 | 4.9 | 4.9 | Debit Card
4 | 2024-10-21 | 17:17:25 | "CNR8906825" | Completed | "CID2610914" | Premier Sedan | Central Secretariat | Inderlok | 13.1 | 28.5 | null | null | null | null | null | null | 416 | 34.02 | 4.6 | 5.0 | UPI
5 | 2024-09-16 | 22:08:00 | "CNR1950162" | Completed | "CID9933542" | Bike | Ghitorni Village | Khan Market | 5.3 | 19.6 | null | null | null | null | null | null | 737 | 48.21 | 4.1 | 4.3 | UPI
6 | 2024-02-06 | 09:44:56 | "CNR4096693" | Completed | "CID4670564" | Auto | AIIMS | Narsinghpur | 5.1 | 18.1 | null | null | null | null | null | null | 316 | 4.85 | 4.1 | 4.6 | UPI
7 | 2024-06-17 | 15:45:58 | "CNR2002539" | Completed | "CID6800553" | Go Mini | Vaishali | Punjabi Bagh | 7.1 | 20.4 | null | null | null | null | null | null | 640 | 41.24 | 4.0 | 4.1 | UPI
8 | 2024-03-19 | 17:37:37 | "CNR6568000" | Completed | "CID8610436" | Auto | Mayur Vihar | Cyber Hub | 12.1 | 16.5 | null | null | null | null | null | null | 136 | 6.56 | 4.4 | 4.2 | UPI
9 | 2024-09-14 | 12:49:09 | "CNR4510807" | No Driver Found | "CID7873618" | Go Sedan | Noida Sector 62 | Noida Sector 18 | null | null | null | null | null | null | null | null | null | null | null | null | null
10 | 2024-12-16 | 19:06:48 | "CNR7721892" | Incomplete | "CID5214275" | Auto | Rohini | Adarsh Nagar | 6.1 | 26.0 | null | null | null | null | 1 | Other Issue | 135 | 10.36 | null | null | Cash
11 | 2024-06-14 | 16:24:12 | "CNR9070334" | Completed | "CID6680340" | Auto | Udyog Bhawan | Dwarka Sector 21 | 7.7 | 18.9 | null | null | null | null | null | null | 181 | 19.84 | 4.2 | 4.9 | Cash
12 | 2024-09-18 | 08:09:38 | "CNR9551927" | No Driver Found | "CID7568143" | Auto | Vidhan Sabha | AIIMS | null | null | null | null | null | null | null | null | null | null | null | null | null
13 | 2024-06-25 | 22:44:15 | "CNR4386945" | Cancelled by Driver | "CID5543520" | eBike | Patel Chowk | Kherki Daula Toll | 4.6 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
14 | 2024-09-11 | 19:29:39 | "CNR2987763" | Completed | "CID2669710" | Go Mini | Malviya Nagar | Ghitorni Village | 12.2 | 28.2 | null | null | null | null | null | null | 394 | 21.44 | 4.1 | 4.7 | UPI
15 | 2024-10-18 | 18:28:53 | "CNR8962232" | Completed | "CID1789354" | Go Mini | Madipur | GTB Nagar | 14.0 | 30.9 | null | null | null | null | null | null | 836 | 39.55 | 4.7 | 4.4 | UPI
16 | 2024-06-07 | 15:05:35 | "CNR2390352" | Completed | "CID5432215" | Auto | Jama Masjid | Khan Market | 8.5 | 36.9 | null | null | null | null | null | null | 410 | 34.76 | 4.0 | 4.9 | Uber Wallet
17 | 2024-07-01 | 10:51:16 | "CNR3221338" | Completed | "CID2581698" | Premier Sedan | IGI Airport | Madipur | 5.6 | 27.5 | null | null | null | null | null | null | 401 | 21.97 | 4.9 | 4.3 | UPI
18 | 2024-12-15 | 15:08:25 | "CNR6739317" | Cancelled by Driver | "CID8682675" | Go Sedan | Vinobapuri | GTB Nagar | 6.0 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
19 | 2024-11-24 | 09:07:10 | "CNR6126048" | Cancelled by Customer | "CID1060329" | eBike | Kashmere Gate | Anand Vihar | 12.4 | null | 1 | Driver is not moving towards pickup location | null | null | null | null | null | null | null
20 | 2024-05-24 | 19:53:57 | "CNR9465840" | Cancelled by Driver | "CID9046501" | eBike | Pitampura | Rajiv Nagar | 10.3 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
21 | 2024-12-14 | 21:27:04 | "CNR9843229" | Completed | "CID4929303" | Go Sedan | Punjabi Bagh | Mansarovar Park | 6.1 | 33.7 | null | null | null | null | null | null | 264 | 13.64 | 3.8 | 4.7 | Uber Wallet
22 | 2024-07-29 | 06:28:22 | "CNR7911434" | Completed | "CID8465004" | Auto | Greater Noida | Botanical Garden | 7.0 | 21.3 | null | null | null | null | null | null | 183 | 4.5 | 5.0 | 4.2 | UPI
23 | 2024-02-23 | 11:02:45 | "CNR6178330" | Completed | "CID7719678" | Go Mini | Tis Hazari | IMT Manesar | 10.4 | 25.1 | null | null | null | null | null | null | 713 | 38.03 | 3.7 | 3.6 | UPI
24 | 2024-03-09 | 20:14:24 | "CNR5309981" | Completed | "CID3870048" | Go Mini | Noida Sector 18 | Old Gurgaon | 7.9 | 41.1 | null | null | null | null | null | null | 113 | 14.53 | 4.6 | 4.9 | UPI
25 | 2024-08-28 | 16:30:14 | "CNR2978950" | Completed | "CID2977484" | Auto | Kanhaiya Nagar | Anand Vihar | 13.6 | 25.2 | null | null | null | null | null | null | 767 | 7.39 | 3.8 | 4.6 | Debit Card
26 | 2024-08-08 | 06:34:38 | "CNR2806943" | Completed | "CID7550481" | Uber XL | Okhla | Barakhamba Road | 5.1 | 20.7 | null | null | null | null | null | null | 495 | 10.23 | 3.8 | 4.2 | Credit Card
27 | 2024-01-26 | 09:44:00 | "CNR3614535" | Cancelled by Driver | "CID2507102" | Auto | Cyber Hub | Saket | 11.5 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
28 | 2024-04-12 | 19:42:35 | "CNR4499383" | No Driver Found | "CID5717521" | Premier Sedan | Sadar Bazar Gurgaon | Mehrauli | null | null | null | null | null | null | null | null | null | null | null | null | null
29 | 2024-04-05 | 18:57:12 | "CNR9834201" | Incomplete | "CID7891331" | Bike | Shastri Park | Botanical Garden | 3.1 | 28.2 | null | null | null | null | 1 | Vehicle Breakdown | 304 | 1.98 | null | null | UPI
30 | 2024-11-03 | 12:24:59 | "CNR8779845" | Completed | "CID7444351" | Auto | Faridabad Sector 15 | Malviya Nagar | 8.9 | 31.1 | null | null | null | null | null | null | 2847 | 39.21 | 3.1 | 4.6 | Debit Card
31 | 2024-05-18 | 17:37:52 | "CNR2178654" | Cancelled by Driver | "CID3631860" | Go Mini | Qutub Minar | Vishwavidyalaya | 9.2 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
32 | 2024-11-15 | 05:53:55 | "CNR3871541" | Completed | "CID9849259" | Go Mini | Mundka | Preet Vihar | 8.9 | 41.0 | null | null | null | null | null | null | 282 | 8.36 | 3.8 | 4.4 | UPI
33 | 2024-02-04 | 18:08:04 | "CNR2877692" | Completed | "CID5901711" | Premier Sedan | DLF City Court | Nehru Place | 11.5 | 20.5 | null | null | null | null | null | null | 656 | 6.39 | 5.0 | 4.9 | Uber Wallet
34 | 2024-05-29 | 14:05:10 | "CNR9603633" | Completed | "CID5856503" | Auto | Tis Hazari | Shahdara | 2.0 | 19.4 | null | null | null | null | null | null | 134 | 46.66 | 3.8 | 4.4 | Credit Card
35 | 2024-07-13 | 19:53:15 | "CNR8087051" | Cancelled by Driver | "CID9716528" | Go Sedan | New Colony | Ghitorni Village | 5.4 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
36 | 2024-06-07 | 19:21:10 | "CNR7823292" | Completed | "CID9377721" | Go Sedan | Nirman Vihar | Malviya Nagar | 14.6 | 25.4 | null | null | null | null | null | null | 470 | 27.07 | 3.7 | 4.2 | Cash
37 | 2024-05-16 | 13:04:31 | "CNR9918895" | Completed | "CID9520433" | Auto | New Delhi Railway Station | Noida Film City | 6.7 | 20.5 | null | null | null | null | null | null | 187 | 12.16 | 4.9 | 4.3 | UPI
38 | 2024-10-02 | 03:10:35 | "CNR7401001" | Completed | "CID9958715" | Premier Sedan | Civil Lines Gurgaon | Mandi House | 6.6 | 30.0 | null | null | null | null | null | null | 186 | 10.28 | 4.4 | 4.6 | UPI
39 | 2024-02-26 | 15:43:16 | "CNR2631310" | Completed | "CID4355670" | eBike | Seelampur | Janakpuri | 10.0 | 35.8 | null | null | null | null | null | null | 70 | 44.42 | 4.1 | 4.4 | UPI
40 | 2024-09-10 | 13:02:42 | "CNR4218487" | Cancelled by Customer | "CID3037053" | Bike | Noida Extension | Udyog Vihar Phase 4 | 11.0 | null | 1 | Driver is not moving towards pickup location | null | null | null | null | null | null | null
41 | 2024-09-12 | 09:14:57 | "CNR5333099" | Cancelled by Driver | "CID5112050" | Auto | Adarsh Nagar | Civil Lines Gurgaon | 8.0 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
42 | 2024-01-24 | 10:47:57 | "CNR5969954" | Completed | "CID5558118" | Auto | Vinobapuri | Karkarduma | 12.8 | 18.7 | null | null | null | null | null | null | 273 | 34.64 | 4.5 | 4.6 | Uber Wallet
43 | 2024-02-12 | 19:42:39 | "CNR4130953" | Incomplete | "CID3196782" | Premier Sedan | Panipat | Tagore Garden | 6.4 | 10.5 | null | null | null | null | 1 | Other Issue | 966 | 18.3 | null | null | Debit Card
44 | 2024-10-02 | 19:17:00 | "CNR6693125" | Cancelled by Driver | "CID7364449" | Auto | Karol Bagh | Tagore Garden | 5.4 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
45 | 2024-04-01 | 21:14:19 | "CNR9572979" | Completed | "CID3474992" | Go Sedan | Sultanpur | Noida Extension | 12.2 | 42.3 | null | null | null | null | null | null | 251 | 20.91 | 4.2 | 3.6 | Debit Card
46 | 2024-09-18 | 21:06:36 | "CNR6444364" | Cancelled by Driver | "CID8262635" | Go Mini | Moti Nagar | Anand Vihar ISBT | 6.6 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
47 | 2024-08-02 | 07:17:07 | "CNR4862806" | Cancelled by Customer | "CID7875150" | Auto | Shastri Park | Anand Vihar ISBT | 11.3 | null | 1 | Driver asked to cancel | null | null | null | null | null | null | null
48 | 2024-09-30 | 13:10:50 | "CNR4674790" | Incomplete | "CID4795905" | Bike | Dilshad Garden | Nehru Place | 2.3 | 29.7 | null | null | null | null | 1 | Vehicle Breakdown | 453 | 19.24 | null | null | Debit Card
49 | 2024-06-16 | 00:44:47 | "CNR6866217" | Completed | "CID7798304" | Go Sedan | Dilshad Garden | Nehru Place | 7.8 | 42.2 | null | null | null | null | null | null | 246 | 8.13 | 4.2 | 4.5 | UPI
50 | 2024-09-18 | 09:33:55 | "CNR4932611" | Completed | "CID8795963" | Auto | Aya Nagar | Karkarduma | 8.8 | 25.2 | null | null | null | null | null | null | 237 | 2.96 | 3.9 | 3.3 | UPI
51 | 2024-05-04 | 21:26:59 | "CNR1572483" | Incomplete | "CID6059557" | Go Mini | Rajiv Chowk | Central Secretariat | 9.9 | 21.6 | null | null | null | null | 1 | Vehicle Breakdown | 1581 | 1.82 | null | null | Cash
52 | 2024-06-03 | 18:14:47 | "CNR6945858" | Completed | "CID9079088" | Go Mini | MG Road | Hauz Rani | 4.1 | 25.1 | null | null | null | null | null | null | 379 | 33.04 | 4.2 | 4.5 | Uber Wallet
53 | 2024-10-03 | 22:20:34 | "CNR6584679" | Completed | "CID4738537" | Go Mini | Jasola | Palam Vihar | 11.2 | 40.7 | null | null | null | null | null | null | 545 | 44.7 | 4.3 | 4.3 | Cash
54 | 2024-05-25 | 11:24:29 | "CNR6125653" | Completed | "CID2808212" | Go Mini | Kashmere Gate | RK Puram | 3.0 | 35.4 | null | null | null | null | null | null | 479 | 7.38 | 4.2 | 4.6 | UPI
55 | 2024-03-04 | 17:46:46 | "CNR6092529" | Cancelled by Driver | "CID5040634" | Go Sedan | Ardee City | Basai Dhankot | 11.2 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
56 | 2024-11-10 | 05:13:57 | "CNR6339107" | Completed | "CID7358763" | Go Mini | Punjabi Bagh | Madipur | 9.1 | 23.2 | null | null | null | null | null | null | 647 | 24.93 | 4.3 | 4.7 | Uber Wallet
57 | 2024-09-11 | 10:36:44 | "CNR1349015" | Cancelled by Driver | "CID9386278" | Auto | Meerut | Badarpur | 4.3 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
58 | 2024-04-11 | 15:43:34 | "CNR9773309" | No Driver Found | "CID9965847" | Uber XL | Anand Vihar ISBT | Dwarka Sector 21 | null | null | null | null | null | null | null | null | null | null | null | null | null
59 | 2024-06-01 | 10:03:27 | "CNR2107544" | Completed | "CID1266188" | Premier Sedan | Mayur Vihar | Palam Vihar | 11.4 | 42.1 | null | null | null | null | null | null | 125 | 25.16 | 4.3 | 5.0 | Credit Card
60 | 2024-05-09 | 10:44:30 | "CNR4171959" | Cancelled by Driver | "CID8470723" | Premier Sedan | Lajpat Nagar | Ramesh Nagar | 6.3 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
61 | 2024-07-30 | 12:02:26 | "CNR5316307" | Completed | "CID1816553" | Premier Sedan | Tughlakabad | Akshardham | 12.4 | 21.5 | null | null | null | null | null | null | 375 | 48.85 | 3.3 | 3.3 | Cash
62 | 2024-02-25 | 16:23:19 | "CNR3635865" | Completed | "CID6477355" | Auto | Cyber Hub | Yamuna Bank | 11.7 | 33.0 | null | null | null | null | null | null | 310 | 24.41 | 4.6 | 3.8 | Debit Card
63 | 2024-02-09 | 11:15:59 | "CNR2497989" | Cancelled by Customer | "CID5007066" | Auto | Karkarduma | IGI Airport | 16.6 | null | 1 | Driver is not moving towards pickup location | null | null | null | null | null | null | null
64 | 2024-03-02 | 04:59:03 | "CNR8946337" | Cancelled by Driver | "CID6438326" | Go Sedan | Shastri Nagar | New Colony | 5.4 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
65 | 2024-08-27 | 18:22:02 | "CNR3740694" | Incomplete | "CID5885514" | Auto | Aya Nagar | Green Park | 2.3 | 10.1 | null | null | null | null | 1 | Vehicle Breakdown | 370 | 13.85 | null | null | Cash
66 | 2024-03-08 | 09:28:38 | "CNR4660896" | Cancelled by Driver | "CID5486947" | Premier Sedan | Dwarka Mor | ITO | 5.8 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
67 | 2024-03-20 | 09:59:11 | "CNR2513330" | Cancelled by Driver | "CID7805280" | Premier Sedan | Adarsh Nagar | Noida Extension | 6.7 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
68 | 2024-01-25 | 12:39:21 | "CNR9252424" | Completed | "CID9413620" | eBike | Vidhan Sabha | Mehrauli | 12.1 | 18.1 | null | null | null | null | null | null | 291 | 3.71 | 3.2 | 4.3 | UPI
69 | 2024-04-11 | 11:14:35 | "CNR3096031" | Completed | "CID3101926" | Auto | Anand Vihar | Saket | 5.8 | 24.3 | null | null | null | null | null | null | 114 | 32.98 | 4.3 | 4.5 | Uber Wallet
70 | 2024-06-17 | 20:25:47 | "CNR4141159" | Completed | "CID6810411" | Go Mini | Uttam Nagar | New Delhi Railway Station | 4.0 | 44.5 | null | null | null | null | null | null | 464 | 11.2 | 4.5 | 4.2 | Cash
71 | 2024-07-29 | 19:52:48 | "CNR6757829" | Completed | "CID8433663" | Go Sedan | Model Town | Mundka | 3.4 | 27.1 | null | null | null | null | null | null | 160 | 45.4 | 4.0 | 5.0 | Cash
72 | 2024-12-22 | 10:53:20 | "CNR6204326" | Cancelled by Driver | "CID9419495" | Auto | Mandi House | India Gate | 11.1 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
73 | 2024-12-21 | 21:40:00 | "CNR3347993" | Completed | "CID4871438" | eBike | Ghitorni | Pitampura | 9.9 | 38.5 | null | null | null | null | null | null | 206 | 28.78 | 4.1 | 3.7 | Cash
74 | 2024-02-23 | 11:38:09 | "CNR6796311" | Incomplete | "CID1880505" | Go Mini | Ghitorni Village | Palam Vihar | 5.9 | 17.2 | null | null | null | null | 1 | Customer Demand | 217 | 2.81 | null | null | Credit Card
75 | 2024-05-05 | 21:22:39 | "CNR3908312" | Completed | "CID8787746" | Bike | Okhla | Netaji Subhash Place | 8.1 | 37.3 | null | null | null | null | null | null | 633 | 17.34 | 4.5 | 4.9 | Credit Card
76 | 2024-11-08 | 17:01:10 | "CNR9416323" | Completed | "CID1025742" | Auto | DLF City Court | Udyog Vihar Phase 4 | 6.1 | 28.2 | null | null | null | null | null | null | 717 | 9.38 | 4.2 | 4.5 | UPI
77 | 2024-12-26 | 21:21:55 | "CNR2912296" | Completed | "CID8094609" | Go Sedan | Rajiv Nagar | Ramesh Nagar | 6.0 | 26.7 | null | null | null | null | null | null | 151 | 11.9 | 4.9 | 4.5 | UPI
78 | 2024-09-05 | 07:44:18 | "CNR4611314" | Completed | "CID6968311" | Premier Sedan | Sohna Road | New Delhi Railway Station | 5.0 | 34.7 | null | null | null | null | null | null | 512 | 35.66 | 3.7 | 4.9 | UPI
79 | 2024-11-25 | 08:29:14 | "CNR2601752" | Cancelled by Customer | "CID9283370" | Go Sedan | Sadar Bazar Gurgaon | Indraprastha | 12.0 | null | 1 | Driver asked to cancel | null | null | null | null | null | null | null
80 | 2024-08-15 | 18:40:03 | "CNR5793527" | Cancelled by Driver | "CID5705499" | Premier Sedan | Anand Vihar ISBT | Rohini | 3.2 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
81 | 2024-02-10 | 09:17:39 | "CNR4487090" | Completed | "CID1210782" | Premier Sedan | Lajpat Nagar | Chirag Delhi | 9.6 | 39.1 | null | null | null | null | null | null | 1390 | 21.79 | 5.0 | 4.9 | UPI
82 | 2024-09-26 | 11:33:51 | "CNR5266935" | Cancelled by Driver | "CID2831708" | Premier Sedan | Bahadurgarh | Pataudi Chowk | 3.9 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
83 | 2024-07-01 | 17:39:56 | "CNR1487489" | Completed | "CID3403699" | Bike | Moolchand | Karol Bagh | 3.9 | 23.0 | null | null | null | null | null | null | 224 | 7.86 | 4.1 | 4.2 | Cash
84 | 2024-10-19 | 16:24:50 | "CNR6351034" | Completed | "CID2038230" | Auto | Uttam Nagar | Vaishali | 8.4 | 31.3 | null | null | null | null | null | null | 708 | 18.04 | 4.6 | 4.8 | UPI
85 | 2024-02-09 | 10:51:01 | "CNR7600870" | Completed | "CID7289498" | Auto | INA Market | IFFCO Chowk | 7.7 | 43.7 | null | null | null | null | null | null | 153 | 15.18 | 4.2 | 3.6 | Uber Wallet
86 | 2024-03-31 | 12:37:12 | "CNR5444321" | Cancelled by Driver | "CID3326946" | Bike | Sushant Lok | Noida Sector 62 | 6.3 | null | null | null | 1 | The customer was coughing/sick | null | null | null | null | null | null | null
87 | 2024-12-11 | 17:05:36 | "CNR8509610" | Completed | "CID8227993" | Bike | IIT Delhi | Mansarovar Park | 5.0 | 35.0 | null | null | null | null | null | null | 72 | 13.38 | 4.7 | 4.2 | Cash
88 | 2024-06-19 | 16:19:41 | "CNR6851549" | Incomplete | "CID9560489" | Go Sedan | Satguru Ram Singh Marg | Punjabi Bagh | 8.9 | 15.6 | null | null | null | null | 1 | Customer Demand | 282 | 11.81 | null | null | UPI
89 | 2024-06-20 | 18:58:36 | "CNR3318044" | No Driver Found | "CID4051792" | Go Mini | Adarsh Nagar | Gwal Pahari | null | null | null | null | null | null | null | null | null | null | null | null | null
90 | 2024-07-10 | 02:23:12 | "CNR3126177" | Completed | "CID1851430" | Premier Sedan | Munirka | Saidulajab | 10.9 | 43.6 | null | null | null | null | null | null | 696 | 30.9 | 4.7 | 3.7 | UPI
91 | 2024-02-06 | 21:27:00 | "CNR6453539" | Incomplete | "CID6999817" | eBike | New Colony | Jasola | 5.7 | 18.0 | null | null | null | null | 1 | Vehicle Breakdown | 936 | 14.91 | null | null | Uber Wallet
92 | 2024-09-30 | 12:34:51 | "CNR7449144" | Completed | "CID8609350" | Auto | IGNOU Road | Tis Hazari | 8.6 | 32.7 | null | null | null | null | null | null | 2169 | 44.25 | 4.4 | 4.3 | Uber Wallet
93 | 2024-09-20 | 13:09:16 | "CNR1970319" | Completed | "CID6085619" | eBike | Meerut | Udyog Bhawan | 9.5 | 24.6 | null | null | null | null | null | null | 136 | 45.46 | 3.2 | 3.8 | Uber Wallet
94 | 2024-04-30 | 20:25:28 | "CNR4415815" | Completed | "CID8351180" | Go Mini | Punjabi Bagh | Saket A Block | 4.6 | 38.9 | null | null | null | null | null | null | 214 | 2.91 | 4.1 | 4.7 | UPI
95 | 2024-06-04 | 20:38:07 | "CNR3899082" | Completed | "CID5467225" | Go Mini | Akshardham | Mayur Vihar | 10.3 | 16.4 | null | null | null | null | null | null | 835 | 37.57 | 4.2 | 4.8 | UPI
96 | 2024-10-30 | 16:29:36 | "CNR3274519" | Cancelled by Driver | "CID3829804" | eBike | Chhatarpur | Gurgaon Sector 29 | 6.0 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
97 | 2024-08-11 | 18:21:32 | "CNR3263958" | Completed | "CID3231090" | Go Sedan | IMT Manesar | Ashram | 2.6 | 36.2 | null | null | null | null | null | null | 705 | 40.18 | 3.9 | 4.8 | UPI
98 | 2024-05-21 | 17:39:58 | "CNR5822542" | No Driver Found | "CID1628370" | Uber XL | Madipur | DLF City Court | null | null | null | null | null | null | null | null | null | null | null | null | null
99 | 2024-10-08 | 08:52:57 | "CNR9652983" | Completed | "CID4773175" | Go Sedan | Azadpur | Khan Market | 2.1 | 35.5 | null | null | null | null | null | null | 480 | 20.22 | 4.3 | 4.8 | Debit Card
100 | 2024-03-26 | 18:59:52 | "CNR2598306" | Cancelled by Customer | "CID5704025" | Premier Sedan | Chhatarpur | Badshahpur | 14.4 | null | 1 | AC is not working | null | null | null | null | null | null | null
101 | 2024-06-24 | 18:04:39 | "CNR3274827" | Completed | "CID3134967" | Go Mini | Cyber Hub | Old Gurgaon | 7.5 | 41.7 | null | null | null | null | null | null | 707 | 44.76 | 4.8 | 4.4 | UPI
102 | 2024-03-28 | 19:20:01 | "CNR9139265" | Cancelled by Driver | "CID4003339" | Bike | Jhilmil | DLF Phase 3 | 9.9 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
103 | 2024-09-27 | 20:16:53 | "CNR8036415" | Completed | "CID9244184" | eBike | Sikanderpur | Greater Noida | 10.0 | 38.0 | null | null | null | null | null | null | 865 | 43.34 | 3.2 | 3.8 | Cash
104 | 2024-04-26 | 15:45:51 | "CNR8433849" | Completed | "CID7659765" | Bike | Mayur Vihar | IGNOU Road | 10.8 | 40.0 | null | null | null | null | null | null | 768 | 5.97 | 4.3 | 4.8 | Cash
105 | 2024-05-19 | 18:47:16 | "CNR2948784" | Completed | "CID7747807" | Bike | Rohini West | Yamuna Bank | 7.0 | 37.0 | null | null | null | null | null | null | 626 | 27.96 | 3.9 | 4.3 | Uber Wallet
106 | 2024-06-05 | 11:16:12 | "CNR1996806" | Completed | "CID2293483" | Go Mini | Mandi House | Bahadurgarh | 9.2 | 40.4 | null | null | null | null | null | null | 443 | 21.19 | 4.2 | 5.0 | UPI
107 | 2024-01-07 | 16:21:31 | "CNR6241612" | Cancelled by Customer | "CID5715130" | Auto | Azadpur | Hauz Rani | 7.9 | null | 1 | Driver asked to cancel | null | null | null | null | null | null | null
108 | 2024-04-19 | 14:14:27 | "CNR5222120" | Completed | "CID2370613" | Go Sedan | Aya Nagar | Chandni Chowk | 9.9 | 28.7 | null | null | null | null | null | null | 226 | 47.63 | 3.6 | 4.1 | UPI
109 | 2024-03-31 | 16:34:57 | "CNR6596872" | Completed | "CID2589193" | Auto | Shastri Nagar | Saket A Block | 12.8 | 30.6 | null | null | null | null | null | null | 449 | 24.4 | 3.3 | 4.1 | UPI
110 | 2024-03-26 | 19:13:05 | "CNR7149058" | Completed | "CID7341591" | Go Mini | Vinobapuri | Shastri Park | 5.9 | 21.1 | null | null | null | null | null | null | 778 | 7.49 | 4.3 | 4.9 | UPI
111 | 2024-06-10 | 15:24:51 | "CNR5726463" | Completed | "CID3407189" | Bike | Rohini West | Yamuna Bank | 14.7 | 18.8 | null | null | null | null | null | null | 361 | 24.44 | 4.3 | 4.8 | UPI
112 | 2024-08-16 | 17:05:19 | "CNR3975551" | Completed | "CID2944423" | Auto | Kashmere Gate ISBT | New Colony | 3.5 | 30.4 | null | null | null | null | null | null | 235 | 9.2 | 4.3 | 3.9 | UPI
113 | 2024-02-04 | 15:32:16 | "CNR8637930" | Completed | "CID8137475" | Go Sedan | Rohini | Botanical Garden | 8.6 | 15.9 | null | null | null | null | null | null | 186 | 28.09 | 4.2 | 4.4 | Cash
114 | 2024-06-24 | 18:46:34 | "CNR9935095" | Incomplete | "CID7359331" | Uber XL | Kherki Daula Toll | Nawada | 3.2 | 21.9 | null | null | null | null | 1 | Other Issue | 544 | 3.37 | null | null | Cash
115 | 2024-03-23 | 11:31:18 | "CNR8893726" | No Driver Found | "CID8335294" | Bike | IFFCO Chowk | Dilshad Garden | null | null | null | null | null | null | null | null | null | null | null | null | null
116 | 2024-07-21 | 10:45:00 | "CNR3975051" | Cancelled by Customer | "CID5797043" | Go Mini | Hauz Rani | Kaushambi | 13.1 | null | 1 | Driver asked to cancel | null | null | null | null | null | null | null
117 | 2024-09-22 | 06:39:39 | "CNR1984811" | Cancelled by Customer | "CID6701334" | Auto | Dilshad Garden | Qutub Minar | 8.1 | null | 1 | Change of plans | null | null | null | null | null | null | null
118 | 2024-05-26 | 22:36:05 | "CNR7036055" | Completed | "CID7359512" | Bike | Faridabad Sector 15 | Vinobapuri | 7.7 | 25.0 | null | null | null | null | null | null | 86 | 40.1 | 3.6 | 4.4 | Cash
119 | 2024-03-30 | 13:55:19 | "CNR3449107" | Cancelled by Driver | "CID8684841" | Auto | Sarai Kale Khan | Central Secretariat | 8.1 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
120 | 2024-09-19 | 08:47:23 | "CNR6919653" | No Driver Found | "CID5736754" | Go Mini | Vishwavidyalaya | MG Road | null | null | null | null | null | null | null | null | null | null | null | null | null
121 | 2024-12-26 | 16:16:56 | "CNR8504461" | Completed | "CID3984123" | Premier Sedan | Malviya Nagar | Chanakyapuri | 8.3 | 43.7 | null | null | null | null | null | null | 234 | 42.5 | 3.7 | 4.9 | Cash
122 | 2024-06-28 | 10:22:29 | "CNR3712671" | Cancelled by Driver | "CID5160108" | eBike | Mayur Vihar | Vidhan Sabha | 11.9 | null | null | null | 1 | The customer was coughing/sick | null | null | null | null | null | null | null
123 | 2024-11-30 | 18:36:50 | "CNR8819329" | Completed | "CID6695773" | Go Sedan | Vaishali | Tilak Nagar | 6.0 | 44.4 | null | null | null | null | null | null | 758 | 41.42 | 4.9 | 3.6 | Cash
124 | 2024-11-04 | 13:46:26 | "CNR3734278" | Incomplete | "CID6462328" | Auto | Kashmere Gate ISBT | Laxmi Nagar | 7.9 | 16.7 | null | null | null | null | 1 | Other Issue | 629 | 17.52 | null | null | UPI
125 | 2024-04-16 | 19:41:40 | "CNR2472864" | Completed | "CID7156852" | Auto | Maidan Garhi | Panchsheel Park | 12.5 | 26.3 | null | null | null | null | null | null | 273 | 31.69 | 4.4 | 4.8 | UPI
126 | 2024-09-12 | 05:52:14 | "CNR3239651" | Completed | "CID9573465" | Auto | Yamuna Bank | Nirman Vihar | 11.1 | 19.6 | null | null | null | null | null | null | 117 | 33.74 | 4.5 | 3.7 | Debit Card
127 | 2024-06-07 | 18:08:10 | "CNR2646387" | Completed | "CID8652330" | Go Sedan | Khan Market | Central Secretariat | 13.3 | 15.8 | null | null | null | null | null | null | 955 | 36.82 | 3.7 | 3.7 | UPI
128 | 2024-06-01 | 15:55:36 | "CNR4724814" | Completed | "CID1251463" | Go Sedan | Pragati Maidan | Jama Masjid | 9.7 | 31.5 | null | null | null | null | null | null | 139 | 38.75 | 4.4 | 4.3 | UPI
129 | 2024-05-05 | 14:17:24 | "CNR2200152" | Completed | "CID6635464" | Go Mini | Mansarovar Park | Rohini | 13.3 | 30.7 | null | null | null | null | null | null | 255 | 17.07 | 4.3 | 4.4 | Cash
130 | 2024-01-15 | 17:05:26 | "CNR5060137" | Completed | "CID2961093" | Auto | Rithala | Aya Nagar | 9.5 | 18.0 | null | null | null | null | null | null | 1212 | 19.18 | 4.3 | 4.3 | UPI
131 | 2024-07-08 | 17:50:10 | "CNR9291876" | Completed | "CID9026677" | Go Sedan | Shastri Park | Indraprastha | 10.0 | 23.6 | null | null | null | null | null | null | 81 | 23.93 | 4.7 | 3.9 | Debit Card
132 | 2024-08-14 | 17:41:00 | "CNR4212658" | Completed | "CID5774800" | Auto | Cyber Hub | Udyog Bhawan | 3.6 | 24.2 | null | null | null | null | null | null | 940 | 29.09 | 4.2 | 4.6 | UPI
133 | 2024-07-31 | 17:07:36 | "CNR7677229" | Cancelled by Driver | "CID8986725" | Auto | Jhilmil | Vidhan Sabha | 8.5 | null | null | null | 1 | The customer was coughing/sick | null | null | null | null | null | null | null
134 | 2024-05-08 | 16:37:21 | "CNR4415643" | Completed | "CID4718297" | eBike | Saidulajab | Hauz Khas | 10.0 | 17.9 | null | null | null | null | null | null | 98 | 27.55 | 4.2 | 4.1 | Cash
135 | 2024-07-29 | 12:28:32 | "CNR4092218" | Completed | "CID5726360" | Premier Sedan | Tughlakabad | Meerut | 5.9 | 15.5 | null | null | null | null | null | null | 724 | 14.0 | 4.1 | 4.2 | Cash
136 | 2024-08-09 | 16:41:35 | "CNR8034979" | Completed | "CID1915443" | eBike | Kashmere Gate | Karkarduma | 8.0 | 25.3 | null | null | null | null | null | null | 749 | 47.3 | 4.4 | 4.3 | UPI
137 | 2024-08-15 | 22:15:54 | "CNR7228014" | No Driver Found | "CID3515824" | eBike | Madipur | Saket | null | null | null | null | null | null | null | null | null | null | null | null | null
138 | 2024-10-02 | 09:10:13 | "CNR1435581" | Completed | "CID1856576" | Auto | Shastri Park | Barakhamba Road | 11.5 | 37.4 | null | null | null | null | null | null | 244 | 14.59 | 4.1 | 4.8 | UPI
139 | 2024-05-12 | 01:09:07 | "CNR5373942" | No Driver Found | "CID7792143" | Auto | Nawada | Jahangirpuri | null | null | null | null | null | null | null | null | null | null | null | null | null
140 | 2024-02-20 | 18:51:34 | "CNR3434703" | Incomplete | "CID3333217" | Auto | Shahdara | Bhiwadi | 2.6 | 22.8 | null | null | null | null | 1 | Vehicle Breakdown | 351 | 8.68 | null | null | UPI
141 | 2024-05-26 | 20:45:58 | "CNR6987620" | Completed | "CID3121337" | Auto | Tilak Nagar | Azadpur | 5.4 | 28.9 | null | null | null | null | null | null | 585 | 34.14 | 4.6 | 4.1 | Debit Card
142 | 2024-12-24 | 10:30:20 | "CNR4132145" | Completed | "CID5140786" | Auto | DLF Phase 3 | Udyog Vihar Phase 4 | 6.4 | 44.7 | null | null | null | null | null | null | 554 | 31.79 | 4.3 | 4.6 | Cash
143 | 2024-06-07 | 17:11:59 | "CNR8207805" | Completed | "CID6685791" | Bike | Laxmi Nagar | Sohna Road | 10.8 | 28.5 | null | null | null | null | null | null | 803 | 25.01 | 4.1 | 4.5 | Credit Card
144 | 2024-04-28 | 19:59:01 | "CNR1438443" | Completed | "CID5468663" | Go Mini | Huda City Centre | Akshardham | 9.4 | 24.8 | null | null | null | null | null | null | 388 | 17.56 | 4.2 | 4.7 | UPI
145 | 2024-01-26 | 18:13:59 | "CNR6129081" | Cancelled by Customer | "CID2998238" | Bike | Shastri Nagar | Shivaji Park | 7.0 | null | 1 | Wrong Address | null | null | null | null | null | null | null
146 | 2024-01-18 | 23:04:30 | "CNR1361929" | Completed | "CID9052724" | Go Mini | Uttam Nagar | Greater Kailash | 11.2 | 38.5 | null | null | null | null | null | null | 192 | 48.96 | 3.1 | 4.6 | UPI
147 | 2024-04-15 | 19:03:28 | "CNR2687237" | Completed | "CID4887972" | Go Mini | Mehrauli | Nawada | 9.0 | 39.1 | null | null | null | null | null | null | 79 | 32.91 | 4.4 | 3.8 | Uber Wallet
148 | 2024-12-19 | 20:42:59 | "CNR7486337" | No Driver Found | "CID8581638" | Go Mini | Mansarovar Park | Manesar | null | null | null | null | null | null | null | null | null | null | null | null | null
149 | 2024-01-22 | 18:23:40 | "CNR4269906" | Completed | "CID9683944" | Bike | Lajpat Nagar | Okhla | 2.4 | 26.6 | null | null | null | null | null | null | 242 | 45.7 | 4.0 | 4.9 | UPI
150 | 2024-08-08 | 21:13:51 | "CNR1224347" | Cancelled by Customer | "CID5493342" | Go Sedan | Patel Chowk | Nirman Vihar | 13.4 | null | 1 | Driver asked to cancel | null | null | null | null | null | null | null
151 | 2024-05-21 | 20:31:57 | "CNR1401386" | Completed | "CID2154950" | Go Mini | Botanical Garden | Azadpur | 5.4 | 32.6 | null | null | null | null | null | null | 323 | 38.32 | 4.4 | 3.7 | Uber Wallet
152 | 2024-08-17 | 12:35:03 | "CNR2013487" | Cancelled by Driver | "CID4175153" | Auto | Bahadurgarh | Pragati Maidan | 7.7 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
153 | 2024-06-02 | 21:27:23 | "CNR7409512" | Completed | "CID5474366" | Auto | Karol Bagh | Palam Vihar | 5.4 | 24.4 | null | null | null | null | null | null | 799 | 9.1 | 4.2 | 4.6 | UPI
154 | 2024-03-11 | 22:20:11 | "CNR7658099" | Cancelled by Customer | "CID8951576" | eBike | Azadpur | Raj Nagar Extension | 19.1 | null | 1 | Change of plans | null | null | null | null | null | null | null
155 | 2024-11-20 | 16:42:58 | "CNR1980045" | Completed | "CID9418411" | Premier Sedan | Ramesh Nagar | Greater Kailash | 2.9 | 39.3 | null | null | null | null | null | null | 134 | 49.26 | 4.1 | 4.9 | UPI
156 | 2024-10-23 | 10:50:53 | "CNR8584951" | Cancelled by Driver | "CID2820460" | Go Mini | Malviya Nagar | Lajpat Nagar | 3.9 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
157 | 2024-08-10 | 18:36:43 | "CNR2607733" | Cancelled by Customer | "CID5637874" | Bike | Dwarka Sector 21 | Malviya Nagar | 18.5 | null | 1 | Change of plans | null | null | null | null | null | null | null
158 | 2024-05-06 | 19:24:18 | "CNR6693847" | Cancelled by Customer | "CID6386344" | Bike | Noida Film City | Keshav Puram | 16.1 | null | 1 | Driver asked to cancel | null | null | null | null | null | null | null
159 | 2024-11-13 | 06:51:43 | "CNR9749627" | Completed | "CID3065827" | Go Sedan | Nirman Vihar | Satguru Ram Singh Marg | 13.7 | 40.7 | null | null | null | null | null | null | 113 | 8.21 | 4.3 | 5.0 | UPI
160 | 2024-09-16 | 16:07:51 | "CNR1706908" | Incomplete | "CID1807270" | Go Mini | Pragati Maidan | Faridabad Sector 15 | 2.3 | 19.6 | null | null | null | null | 1 | Vehicle Breakdown | 84 | 16.09 | null | null | Uber Wallet
161 | 2024-05-16 | 12:35:33 | "CNR6236514" | Completed | "CID3761463" | Auto | Tughlakabad | Ramesh Nagar | 12.1 | 32.5 | null | null | null | null | null | null | 81 | 11.94 | 4.0 | 4.5 | UPI
162 | 2024-02-15 | 19:11:02 | "CNR9048385" | Completed | "CID5242973" | Go Sedan | India Gate | Kashmere Gate | 4.5 | 43.1 | null | null | null | null | null | null | 992 | 8.23 | 5.0 | 4.6 | UPI
163 | 2024-08-01 | 06:16:08 | "CNR1903168" | Completed | "CID1051794" | Premier Sedan | Maidan Garhi | Samaypur Badli | 3.6 | 26.5 | null | null | null | null | null | null | 513 | 49.5 | 4.3 | 4.9 | UPI
164 | 2024-05-02 | 19:16:35 | "CNR2922907" | Completed | "CID4507250" | Premier Sedan | Barakhamba Road | Noida Film City | 5.8 | 38.9 | null | null | null | null | null | null | 635 | 37.37 | 4.3 | 4.6 | UPI
165 | 2024-08-27 | 06:04:30 | "CNR7621341" | Completed | "CID1429306" | Uber XL | Kashmere Gate | Jor Bagh | 6.9 | 17.0 | null | null | null | null | null | null | 242 | 35.38 | 4.4 | 4.9 | UPI
166 | 2024-10-20 | 12:47:05 | "CNR5803828" | Completed | "CID1039074" | Bike | Barakhamba Road | DLF City Court | 12.5 | 20.6 | null | null | null | null | null | null | 1114 | 34.81 | 4.1 | 4.2 | Uber Wallet
167 | 2024-06-25 | 23:08:22 | "CNR4868713" | Completed | "CID2089650" | Bike | Gurgaon Sector 56 | Cyber Hub | 9.0 | 37.7 | null | null | null | null | null | null | 372 | 20.92 | 3.8 | 4.6 | UPI
168 | 2024-02-13 | 06:11:32 | "CNR2575334" | Cancelled by Driver | "CID7330180" | Bike | Ghitorni Village | Faridabad Sector 15 | 6.2 | null | null | null | 1 | The customer was coughing/sick | null | null | null | null | null | null | null
169 | 2024-07-10 | 16:51:38 | "CNR3841045" | Completed | "CID1210899" | Go Mini | Ashram | Yamuna Bank | 3.5 | 40.2 | null | null | null | null | null | null | 387 | 23.39 | 4.3 | 4.7 | Uber Wallet
170 | 2024-06-10 | 07:38:34 | "CNR5265040" | Completed | "CID3076700" | Go Sedan | Gurgaon Sector 29 | Mansarovar Park | 7.1 | 31.6 | null | null | null | null | null | null | 767 | 30.82 | 4.6 | 4.9 | UPI
171 | 2024-07-11 | 14:42:27 | "CNR8112878" | Completed | "CID4809063" | Auto | Shahdara | Paharganj | 7.1 | 24.4 | null | null | null | null | null | null | 447 | 21.31 | 4.5 | 4.1 | UPI
172 | 2024-06-20 | 07:14:32 | "CNR4617937" | Completed | "CID7348314" | Premier Sedan | Sultanpur | Delhi Gate | 6.7 | 33.5 | null | null | null | null | null | null | 500 | 20.41 | 4.3 | 4.7 | Credit Card
173 | 2024-11-18 | 16:38:12 | "CNR7311169" | Cancelled by Driver | "CID2529885" | eBike | Noida Film City | Vasant Kunj | 7.2 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
174 | 2024-03-05 | 12:21:00 | "CNR7340396" | Completed | "CID7334750" | Go Mini | Gurgaon Sector 29 | Dilshad Garden | 3.7 | 38.9 | null | null | null | null | null | null | 446 | 25.94 | 4.9 | 4.3 | Debit Card
175 | 2024-01-12 | 09:19:16 | "CNR4259279" | Cancelled by Driver | "CID7993999" | eBike | Jor Bagh | Golf Course Road | 11.6 | null | null | null | 1 | The customer was coughing/sick | null | null | null | null | null | null | null
176 | 2024-05-17 | 01:08:21 | "CNR7554746" | Completed | "CID3234126" | eBike | Huda City Centre | Aya Nagar | 2.6 | 32.9 | null | null | null | null | null | null | 125 | 39.16 | 4.1 | 4.9 | UPI
177 | 2024-05-13 | 21:08:35 | "CNR9430550" | Completed | "CID3901448" | Go Sedan | Dilshad Garden | Subhash Nagar | 11.4 | 17.2 | null | null | null | null | null | null | 419 | 23.79 | 4.3 | 4.7 | Uber Wallet
178 | 2024-01-04 | 06:58:58 | "CNR9984761" | Cancelled by Driver | "CID1201304" | Go Sedan | Saket | Old Gurgaon | 3.3 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
179 | 2024-06-19 | 22:02:03 | "CNR4112922" | Completed | "CID8248967" | Go Sedan | Basai Dhankot | Qutub Minar | 4.7 | 17.7 | null | null | null | null | null | null | 477 | 41.93 | 4.6 | 4.9 | Cash
180 | 2024-07-01 | 18:52:02 | "CNR5939223" | Cancelled by Driver | "CID1109267" | Go Sedan | Basai Dhankot | Mandi House | 5.8 | null | null | null | 1 | The customer was coughing/sick | null | null | null | null | null | null | null
181 | 2024-11-05 | 17:11:41 | "CNR7645307" | Cancelled by Driver | "CID6238168" | Go Sedan | Jor Bagh | Sadar Bazar Gurgaon | 8.0 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
182 | 2024-04-26 | 21:03:12 | "CNR4722464" | Completed | "CID7557487" | Auto | Anand Vihar | IMT Manesar | 11.3 | 33.0 | null | null | null | null | null | null | 352 | 6.71 | 4.2 | 4.7 | UPI
183 | 2024-02-22 | 22:48:26 | "CNR3191144" | Cancelled by Driver | "CID5636263" | Bike | Ashram | Bhiwadi | 11.1 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
184 | 2024-12-04 | 14:09:18 | "CNR4720063" | Completed | "CID1461597" | Go Sedan | Munirka | Sushant Lok | 7.4 | 39.0 | null | null | null | null | null | null | 117 | 15.63 | 4.6 | 4.5 | UPI
185 | 2024-11-10 | 09:58:14 | "CNR6216825" | Cancelled by Driver | "CID3685295" | Go Sedan | Vishwavidyalaya | Bhikaji Cama Place | 9.4 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
186 | 2024-12-12 | 12:01:07 | "CNR5314934" | Completed | "CID1026547" | Go Sedan | Tis Hazari | Arjangarh | 9.3 | 28.5 | null | null | null | null | null | null | 51 | 29.8 | 4.6 | 3.9 | UPI
187 | 2024-03-02 | 22:28:52 | "CNR5517193" | Completed | "CID9143442" | Uber XL | Malviya Nagar | Sohna Road | 6.0 | 26.6 | null | null | null | null | null | null | 81 | 5.59 | 4.9 | 4.2 | Cash
188 | 2024-07-12 | 16:39:57 | "CNR8412946" | Cancelled by Driver | "CID7723858" | Go Mini | Nawada | Hero Honda Chowk | 8.2 | null | null | null | 1 | The customer was coughing/sick | null | null | null | null | null | null | null
189 | 2024-12-17 | 19:11:24 | "CNR1867567" | Incomplete | "CID6010303" | Premier Sedan | Qutub Minar | Ghitorni | 7.2 | 18.4 | null | null | null | null | 1 | Customer Demand | 635 | 6.57 | null | null | UPI
190 | 2024-08-29 | 21:01:54 | "CNR3455701" | Completed | "CID9643714" | Bike | Munirka | Tis Hazari | 13.5 | 41.8 | null | null | null | null | null | null | 51 | 4.44 | 4.4 | 4.7 | Uber Wallet
191 | 2024-03-25 | 18:40:11 | "CNR5149854" | Completed | "CID9975388" | Go Sedan | Akshardham | IGNOU Road | 4.9 | 33.0 | null | null | null | null | null | null | 469 | 26.93 | 4.3 | 4.9 | UPI
192 | 2024-06-30 | 21:34:48 | "CNR6027511" | Completed | "CID6080643" | Bike | Udyog Vihar Phase 4 | Greater Kailash | 9.9 | 22.0 | null | null | null | null | null | null | 320 | 28.78 | 4.2 | 4.5 | UPI
193 | 2024-07-18 | 02:41:57 | "CNR4569804" | Completed | "CID8474448" | Go Mini | Raj Nagar Extension | IMT Manesar | 13.3 | 24.6 | null | null | null | null | null | null | 726 | 12.98 | 4.2 | 4.3 | Debit Card
194 | 2024-03-14 | 23:23:23 | "CNR7585582" | Incomplete | "CID1607386" | Auto | Ghitorni | Noida Film City | 2.0 | 10.3 | null | null | null | null | 1 | Customer Demand | 68 | 19.06 | null | null | Cash
195 | 2024-09-26 | 12:23:13 | "CNR5534901" | Incomplete | "CID9231793" | eBike | Akshardham | Keshav Puram | 4.4 | 15.7 | null | null | null | null | 1 | Vehicle Breakdown | 136 | 14.0 | null | null | Debit Card
196 | 2024-09-15 | 20:23:02 | "CNR5784218" | Completed | "CID1814194" | Go Sedan | Nawada | New Delhi Railway Station | 8.8 | 16.8 | null | null | null | null | null | null | 455 | 43.06 | 3.8 | 4.3 | UPI
197 | 2024-05-12 | 10:16:09 | "CNR6349771" | Completed | "CID5577961" | Auto | Lajpat Nagar | Shivaji Park | 2.6 | 22.2 | null | null | null | null | null | null | 995 | 38.96 | 4.7 | 4.9 | Debit Card
198 | 2024-11-16 | 19:05:30 | "CNR3708287" | Completed | "CID2958217" | Uber XL | Aya Nagar | Meerut | 2.7 | 25.8 | null | null | null | null | null | null | 1179 | 18.87 | 4.9 | 4.6 | Uber Wallet
199 | 2024-07-19 | 16:57:27 | "CNR8365625" | Completed | "CID4463605" | Bike | Lok Kalyan Marg | Maidan Garhi | 11.0 | 33.1 | null | null | null | null | null | null | 525 | 47.88 | 4.2 | 4.6 | Uber Wallet
200 | 2024-12-02 | 20:04:25 | "CNR7435840" | Cancelled by Driver | "CID9084957" | Auto | Nawada | Pulbangash | 11.3 | null | null | null | 1 | Personal & Car related issues | null | null | null | null | null | null | null
201 | 2024-08-08 | 14:29:16 | "CNR8577529" | Cancelled by Driver | "CID8121948" | Auto | Punjabi Bagh | Badshahpur | 6.9 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
202 | 2024-09-15 | 22:06:40 | "CNR6158921" | Completed | "CID4203688" | Go Sedan | Ramesh Nagar | Keshav Puram | 5.1 | 27.9 | null | null | null | null | null | null | 300 | 5.7 | 4.2 | 4.6 | UPI
203 | 2024-05-30 | 18:57:34 | "CNR7484371" | Completed | "CID7265108" | Go Sedan | Maidan Garhi | Sarai Kale Khan | 11.1 | 18.4 | null | null | null | null | null | null | 202 | 4.62 | 3.2 | 4.1 | UPI
204 | 2024-12-05 | 07:56:34 | "CNR9301716" | Completed | "CID4647277" | Premier Sedan | Greater Kailash | Green Park | 6.3 | 34.1 | null | null | null | null | null | null | 475 | 42.26 | 4.3 | 4.2 | Cash
205 | 2024-07-16 | 19:23:06 | "CNR6511558" | Completed | "CID5808222" | Go Sedan | Sarojini Nagar | Saidulajab | 6.1 | 17.0 | null | null | null | null | null | null | 833 | 19.49 | 4.0 | 3.4 | Debit Card
206 | 2024-06-14 | 16:59:15 | "CNR1801249" | Cancelled by Driver | "CID2012161" | Go Sedan | Satguru Ram Singh Marg | Kashmere Gate | 3.9 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
207 | 2024-01-06 | 20:53:36 | "CNR5712520" | Completed | "CID8152575" | Go Mini | Palam Vihar | Shahdara | 7.3 | 33.1 | null | null | null | null | null | null | 158 | 40.36 | 3.7 | 4.9 | UPI
208 | 2024-02-28 | 03:21:43 | "CNR1038498" | Completed | "CID6242930" | Bike | GTB Nagar | Palam Vihar | 10.0 | 36.9 | null | null | null | null | null | null | 155 | 42.59 | 4.7 | 3.2 | Cash
209 | 2024-04-26 | 18:49:22 | "CNR2395349" | Cancelled by Driver | "CID3389825" | Auto | Vatika Chowk | Pulbangash | 9.3 | null | null | null | 1 | More than permitted people in there | null | null | null | null | null | null | null
210 | 2024-03-31 | 08:32:26 | "CNR4896767" | Completed | "CID6418422" | Auto | Cyber Hub | Mundka | 10.5 | 33.9 | null | null | null | null | null | null | 107 | 21.68 | 4.3 | 4.9 | UPI
211 | 2024-04-04 | 16:33:57 | "CNR5047025" | Completed | "CID6365387" | Go Mini | Adarsh Nagar | Rohini West | 6.8 | 22.4 | null | null | null | null | null | null | 151 | 29.9 | 4.6 | 4.6 | UPI
212 | 2024-05-07 | 21:10:50 | "CNR7514462" | Completed | "CID5973257" | Go Mini | Golf Course Road | Bhikaji Cama Place | 10.7 | 30.4 | null | null | null | null | null | null | 187 | 42.27 | 4.3 | 4.8 | Cash
213 | 2024-06-15 | 19:12:37 | "CNR5269641" | Completed | "CID8225890" | Go Mini | IIT Delhi | Basai Dhankot | 12.3 | 34.6 | null | null | null | null | null | null | 1543 | 49.71 | 4.4 | 5.0 | Cash
214 | 2024-06-14 | 20:58:12 | "CNR9927786" | Completed | "CID8378909" | eBike | Punjabi Bagh | Gurgaon Sector 56 | 3.8 | 20.2 | null | null | null | null | null | null | 432 | 32.56 | 4.2 | 5.0 | UPI
215 | 2024-03-04 | 21:24:16 | "CNR9619867" | Cancelled by Customer | "CID9032039" | Go Sedan | Civil Lines Gurgaon | Moti Nagar | 10.2 | null | 1 | Driver is not moving towards pickup location | null | null | null | null | null | null | null
216 | 2024-01-23 | 10:28:09 | "CNR7250271" | Incomplete | "CID9448818" | Go Sedan | Udyog Vihar Phase 4 | Anand Vihar ISBT | 4.4 | 23.6 | null | null | null | null | 1 | Vehicle Breakdown | 195 | 14.49 | null | null | Uber Wallet
217 | 2024-10-28 | 15:46:27 | "CNR5629774" | Completed | "CID8747096" | Go Mini | Ghaziabad | Udyog Vihar | 3.2 | 30.8 | null | null | null | null | null | null | 54 | 31.49 | 4.6 | 4.9 | Cash
218 | 2024-02-06 | 10:17:43 | "CNR4578797" | Completed | "CID1104786" | Auto | Narsinghpur | Maidan Garhi | 13.4 | 20.0 | null | null | null | null | null | null | 939 | 21.28 | 3.8 | 5.0 | UPI
219 | 2024-12-04 | 07:48:07 | "CNR5277125" | Completed | "CID7166980" | Auto | MG Road | GTB Nagar | 9.8 | 43.4 | null | null | null | null | null | null | 829 | 34.14 | 4.6 | 4.2 | UPI
220 | 2024-12-22 | 17:53:25 | "CNR2343017" | Completed | "CID9190683" | Auto | Panchsheel Park | Gwal Pahari | 8.7 | 27.2 | null | null | null | null | null | null | 314 | 20.66 | 4.6 | 4.3 | UPI
221 | 2024-08-24 | 21:09:54 | "CNR4356773" | Completed | "CID9031824" | Go Sedan | Badarpur | Tis Hazari | 11.7 | 30.5 | null | null | null | null | null | null | 1249 | 24.6 | 4.1 | 3.7 | Cash
222 | 2024-09-18 | 08:48:36 | "CNR1050716" | Completed | "CID3789359" | Auto | Gurgaon Sector 29 | South Extension | 4.5 | 27.4 | null | null | null | null | null | null | 437 | 27.94 | 3.9 | 4.6 | UPI
223 | 2024-10-22 | 08:41:41 | "CNR7850961" | Completed | "CID5103518" | Go Mini | Dwarka Sector 21 | Moolchand | 9.6 | 28.1 | null | null | null | null | null | null | 83 | 44.57 | 4.6 | 4.5 | UPI
224 | 2024-06-16 | 06:53:10 | "CNR1254363" | Completed | "CID9092173" | Auto | Greater Kailash | Raj Nagar Extension | 14.2 | 15.7 | null | null | null | null | null | null | 1444 | 15.43 | 4.2 | 3.1 | Cash
225 | 2024-12-22 | 09:03:55 | "CNR6569789" | Cancelled by Driver | "CID3372949" | Go Mini | Pataudi Chowk | Meerut | 6.5 | null | null | null | 1 | Customer related issue | null | null | null | null | null | null | null
226 | 2024-07-02 | 15:25:18 | "CNR6092310" | No Driver Found | "CID4998701" | Premier Sedan | Bahadurgarh | Uttam Nagar | null | null | null | null | null | null | null | null | null | null | null | null | null
227 | 2024-09-07 | 04:07:03 | "CNR2742692" | Completed | "CID6825812" | Bike | Gwal Pahari | Golf Course Road | 5.8 | 31.4 | null | null | null | null | null | null | 1368 | 42.88 | 4.3 | 3.8 | Cash
228 | 2024-07-05 | 06:41:29 | "CNR7165550" | No Driver Found | "CID7694379" | Bike | Punjabi Bagh | Ghitorni Village | null | null | null | null | null | null | null | null | null | null | null | null | null
229 | 2024-05-03 | 21:53:28 | "CNR9101826" | Completed | "CID1875498" | Auto | Tagore Garden | Narsinghpur | 7.1 | 18.0 | null | null | null | null | null | null | 418 | 23.31 | 4.2 | 4.6 | UPI
230 | 2024-04-10 | 19:13:34 | "CNR7054338" | Completed | "CID5302286" | Premier Sedan | Samaypur Badli | IGNOU Road | 11.0 | 40.0 | null | null | null | null | null | null | 363 | 31.36 | 4.2 | 4.3 | UPI
231 | 2024-02-06 | 09:18:36 | "CNR5871454" | Completed | "CID8316899" | Go Mini | AIIMS | Chandni Chowk | 12.8 | 15.3 | null | null | null | null | null | null | 279 | 16.68 | 4.1 | 4.9 | Cash
232 | 2024-04-22 | 17:13:32 | "CNR4399523" | Completed | "CID2089022" | Go Mini | Hero Honda Chowk | AIIMS | 5.1 | 39.9 | null | null | null | null | null | null | 332 | 49.06 | 4.4 | 4.6 | Cash
233 | 2024-09-25 | 18:58:35 | "CNR9707820" | Completed | "CID2183387" | Auto | Delhi Gate | Gurgaon Sector 29 | 4.7 | 37.9 | null | null | null | null | null | null | 448 | 28.34 | 4.3 | 4.9 | UPI
234 | 2024-07-08 | 18:18:13 | "CNR2693001" | Cancelled by Customer | "CID1889689" | Bike | Rohini West | Peeragarhi | 11.2 | null | 1 | Wrong Address | null | null | null | null | null | null | null
235 | 2024-06-04 | 15:45:01 | "CNR5804667" | Completed | "CID8556234" | Auto | Connaught Place | Gurgaon Sector 56 | 8.3 | 24.7 | null | null | null | null | null | null | 512 | 18.45 | 4.5 | 4.7 | UPI
    """

    # Process the data string
    lines = data.strip().split('\n')
    rows = []
    for line in lines:
        parts = line.split(' | ')
        rows.append(parts)

    # Create DataFrame
    df = pd.DataFrame(rows[1:], columns=rows[0])

    # Clean up the data
    # Remove quotes from string values
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].str.strip('"')

    # Convert numeric columns
    numeric_cols = ['Avg VTAT', 'Avg CTAT', 'Cancelled Rides by Customer',
                   'Cancelled Rides by Driver', 'Incomplete Rides',
                   'Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating']

    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Convert date and time
    df['Date'] = pd.to_datetime(df['Date'])
    df['Time'] = pd.to_datetime(df['Time']).dt.time

    # Extract hour from time for analysis
    df['Hour'] = pd.to_datetime(df['Time'].astype(str)).dt.hour

    # Create a new column for day of week
    df['DayOfWeek'] = df['Date'].dt.day_name()

    # Create a new column for month
    df['Month'] = df['Date'].dt.month_name()

    # Create a new column for ride status category
    def categorize_status(status):
        if status in ['Completed']:
            return 'Successful'
        elif status in ['Cancelled by Customer', 'Cancelled by Driver']:
            return 'Cancelled'
        elif status in ['Incomplete']:
            return 'Incomplete'
        else:
            return 'No Driver Found'

    df['StatusCategory'] = df['Booking Status'].apply(categorize_status)

    # Create a new column for distance category
    def categorize_distance(distance):
        if pd.isna(distance):
            return 'Unknown'
        elif distance < 5:
            return 'Short'
        elif distance < 15:
            return 'Medium'
        else:
            return 'Long'

    df['DistanceCategory'] = df['Ride Distance'].apply(categorize_distance)

    # Create a new column for time of day
    def categorize_time(hour):
        if 5 <= hour < 12:
            return 'Morning'
        elif 12 <= hour < 17:
            return 'Afternoon'
        elif 17 <= hour < 22:
            return 'Evening'
        else:
            return 'Night'

    df['TimeOfDay'] = df['Hour'].apply(categorize_time)

    return df

# Exploratory Data Analysis
def perform_eda(df):
    # Basic statistics
    print("Dataset Shape:", df.shape)
    print("\nData Types:")
    print(df.dtypes)
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Booking status distribution
    plt.figure(figsize=(10, 6))
    status_counts = df['Booking Status'].value_counts()
    plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Booking Status Distribution')
    plt.axis('equal')
    plt.tight_layout()
    plt.savefig('booking_status_distribution.png')
    plt.close()

    # Vehicle type distribution
    plt.figure(figsize=(12, 6))
    vehicle_counts = df['Vehicle Type'].value_counts()
    sns.barplot(x=vehicle_counts.index, y=vehicle_counts.values)
    plt.title('Vehicle Type Distribution')
    plt.xlabel('Vehicle Type')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('vehicle_type_distribution.png')
    plt.close()

    # Booking status by vehicle type
    plt.figure(figsize=(12, 8))
    vehicle_status = pd.crosstab(df['Vehicle Type'], df['Booking Status'])
    vehicle_status.plot(kind='bar', stacked=True)
    plt.title('Booking Status by Vehicle Type')
    plt.xlabel('Vehicle Type')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title='Booking Status')
    plt.tight_layout()
    plt.savefig('booking_status_by_vehicle.png')
    plt.close()

    # Hourly distribution of rides
    plt.figure(figsize=(12, 6))
    hourly_counts = df['Hour'].value_counts().sort_index()
    sns.lineplot(x=hourly_counts.index, y=hourly_counts.values)
    plt.title('Hourly Distribution of Rides')
    plt.xlabel('Hour of Day')
    plt.ylabel('Count')
    plt.xticks(range(0, 24))
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('hourly_distribution.png')
    plt.close()

    # Day of week distribution
    plt.figure(figsize=(10, 6))
    day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    day_counts = df['DayOfWeek'].value_counts().reindex(day_order)
    sns.barplot(x=day_counts.index, y=day_counts.values)
    plt.title('Day of Week Distribution')
    plt.xlabel('Day of Week')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.savefig('day_of_week_distribution.png')
    plt.close()

    # Payment method distribution
    plt.figure(figsize=(10, 6))
    payment_counts = df['Payment Method'].value_counts()
    sns.barplot(x=payment_counts.index, y=payment_counts.values)
    plt.title('Payment Method Distribution')
    plt.xlabel('Payment Method')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('payment_method_distribution.png')
    plt.close()

    # Distance distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Ride Distance'].dropna(), bins=30, kde=True)
    plt.title('Ride Distance Distribution')
    plt.xlabel('Distance (km)')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig('distance_distribution.png')
    plt.close()

    # Booking value distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Booking Value'].dropna(), bins=30, kde=True)
    plt.title('Booking Value Distribution')
    plt.xlabel('Booking Value (INR)')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig('booking_value_distribution.png')
    plt.close()

    # Cancellation reasons
    cancellation_reasons = pd.concat([
        df['Reason for cancelling by Customer'].dropna(),
        df['Driver Cancellation Reason'].dropna(),
        df['Incomplete Rides Reason'].dropna()
    ]).value_counts()

    plt.figure(figsize=(12, 8))
    sns.barplot(x=cancellation_reasons.values, y=cancellation_reasons.index)
    plt.title('Cancellation and Incompletion Reasons')
    plt.xlabel('Count')
    plt.ylabel('Reason')
    plt.tight_layout()
    plt.savefig('cancellation_reasons.png')
    plt.close()

    # Ratings distribution
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    sns.histplot(df['Driver Ratings'].dropna(), bins=10, kde=True, ax=ax1)
    ax1.set_title('Driver Ratings Distribution')
    ax1.set_xlabel('Rating')
    ax1.set_ylabel('Frequency')

    sns.histplot(df['Customer Rating'].dropna(), bins=10, kde=True, ax=ax2)
    ax2.set_title('Customer Rating Distribution')
    ax2.set_xlabel('Rating')
    ax2.set_ylabel('Frequency')

    plt.tight_layout()
    plt.savefig('ratings_distribution.png')
    plt.close()

    # Time of day vs booking status
    plt.figure(figsize=(12, 6))
    time_status = pd.crosstab(df['TimeOfDay'], df['StatusCategory'])
    time_status.plot(kind='bar', stacked=True)
    plt.title('Booking Status by Time of Day')
    plt.xlabel('Time of Day')
    plt.ylabel('Count')
    plt.legend(title='Status')
    plt.tight_layout()
    plt.savefig('time_of_day_status.png')
    plt.close()

    # Distance category vs booking status
    plt.figure(figsize=(12, 6))
    distance_status = pd.crosstab(df['DistanceCategory'], df['StatusCategory'])
    distance_status.plot(kind='bar', stacked=True)
    plt.title('Booking Status by Distance Category')
    plt.xlabel('Distance Category')
    plt.ylabel('Count')
    plt.legend(title='Status')
    plt.tight_layout()
    plt.savefig('distance_category_status.png')
    plt.close()

# Hypothesis Testing
def hypothesis_testing(df):
    # Hypothesis 1: Is there a significant difference in cancellation rates between vehicle types?
    print("\nHypothesis 1: Is there a significant difference in cancellation rates between vehicle types?")

    # Create a contingency table
    vehicle_cancellation = pd.crosstab(df['Vehicle Type'], df['StatusCategory'])
    print("Contingency Table:")
    print(vehicle_cancellation)

    # Chi-square test
    chi2, p, dof, expected = stats.chi2_contingency(vehicle_cancellation)
    print(f"\nChi-square statistic: {chi2}")
    print(f"P-value: {p}")

    if p < 0.05:
        print("Result: Reject the null hypothesis. There is a significant difference in cancellation rates between vehicle types.")
    else:
        print("Result: Fail to reject the null hypothesis. There is no significant difference in cancellation rates between vehicle types.")

    # Hypothesis 2: Is there a significant difference in ride distance between successful and cancelled rides?
    print("\nHypothesis 2: Is there a significant difference in ride distance between successful and cancelled rides?")

    # Extract distance data for successful and cancelled rides
    successful_distances = df[df['StatusCategory'] == 'Successful']['Ride Distance'].dropna()
    cancelled_distances = df[df['StatusCategory'] == 'Cancelled']['Ride Distance'].dropna()

    print(f"Successful rides - Mean distance: {successful_distances.mean():.2f} km, Count: {len(successful_distances)}")
    print(f"Cancelled rides - Mean distance: {cancelled_distances.mean():.2f} km, Count: {len(cancelled_distances)}")

    # T-test
    t_stat, p_val = stats.ttest_ind(successful_distances, cancelled_distances, equal_var=False)
    print(f"\nT-statistic: {t_stat}")
    print(f"P-value: {p_val}")

    if p_val < 0.05:
        print("Result: Reject the null hypothesis. There is a significant difference in ride distance between successful and cancelled rides.")
    else:
        print("Result: Fail to reject the null hypothesis. There is no significant difference in ride distance between successful and cancelled rides.")

    # Hypothesis 3: Is there a significant difference in booking values between payment methods?
    print("\nHypothesis 3: Is there a significant difference in booking values between payment methods?")

    # Extract booking value data for different payment methods
    payment_methods = df['Payment Method'].unique()
    payment_values = []
    for method in payment_methods:
        values = df[df['Payment Method'] == method]['Booking Value'].dropna()
        payment_values.append(values)
        print(f"{method} - Mean booking value: {values.mean():.2f} INR, Count: {len(values)}")

    # ANOVA test
    f_stat, p_val_anova = stats.f_oneway(*payment_values)
    print(f"\nF-statistic: {f_stat}")
    print(f"P-value: {p_val_anova}")

    if p_val_anova < 0.05:
        print("Result: Reject the null hypothesis. There is a significant difference in booking values between payment methods.")
    else:
        print("Result: Fail to reject the null hypothesis. There is no significant difference in booking values between payment methods.")

# Predictive Modeling
def predictive_modeling(df):
    # Prepare data for modeling
    # Select relevant features
    features = ['Vehicle Type', 'DistanceCategory', 'TimeOfDay', 'Payment Method', 'Hour', 'DayOfWeek']
    target = 'StatusCategory'

    # Drop rows with missing target values
    model_df = df.dropna(subset=[target])

    # Encode categorical variables
    le_dict = {}
    for col in features:
        le = LabelEncoder()
        model_df[col] = le.fit_transform(model_df[col].astype(str))
        le_dict[col] = le

    # Encode target variable
    target_le = LabelEncoder()
    model_df[target] = target_le.fit_transform(model_df[target])

    # Split data into train and test sets
    X = model_df[features]
    y = model_df[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Train a Random Forest classifier
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)

    # Make predictions
    y_pred = rf_model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nModel Accuracy: {accuracy:.4f}")

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=target_le.classes_))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_le.classes_, yticklabels=target_le.classes_)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.savefig('confusion_matrix.png')
    plt.close()

    # Feature importance
    feature_importance = pd.DataFrame({
        'Feature': features,
        'Importance': rf_model.feature_importances_
    }).sort_values('Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=feature_importance)
    plt.title('Feature Importance')
    plt.tight_layout()
    plt.savefig('feature_importance.png')
    plt.close()

    return rf_model, le_dict, target_le

# Main function
def main():
    # Load and preprocess data
    df = load_and_preprocess()

    # Perform exploratory data analysis
    perform_eda(df)

    # Hypothesis testing
    hypothesis_testing(df)

    # Predictive modeling
    model, le_dict, target_le = predictive_modeling(df)

    print("\nAnalysis complete. Visualizations saved as PNG files.")

if __name__ == "__main__":
    main()
```
